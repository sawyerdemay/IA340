```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('s3://sawyerandrory/Tokyo Medals 2021.csv')

# Sort the data by 'Rank By Total'
df_sorted = df.sort_values('Rank By Total')

# Set the style for better visualization
plt.style.use('ggplot')

# Create a bar chart
plt.figure(figsize=(20,10))
plt.barh(df_sorted['Country'], df_sorted['Total'], color = 'blue')

# Add labels and title
plt.xlabel('Total Medals')
plt.ylabel('Country')
plt.title('Rank By Total Medals For Each Country in Tokyo 2021 Olympics')

# Display the chart
plt.show()
```

## Comparing the average number of medal rank per country, based on their total amount

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Filter the DataFrame to include only rows where 'Gold Medal' is greater than 0
df_gold = df[df['Gold Medal'] > 0]

# Calculate the average 'Total' for each country that has won at least one gold medal
average_total_by_country = df_gold.groupby('Country')['Total'].mean()

# Create a bar chart
plt.figure(figsize=(10,5))
average_total_by_country.plot(kind='bar', color='gold')

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Average Total Medals')
plt.title('Average Total Medals For Countries With Gold Medals in Tokyo 2021 Olympics')

# Display the chart
plt.show()
```
